XOR (Simétrica)
é um processo simples de criptografia que ele mutiplica bit a bit pela chave escolhida. e o contrario seria a a conversao de bit para texto puro multiplicado pela chave.

Força Bruta
é o processo de testa manualmente a descritografia da mensagem criptgrafada por varias chaves possiveis quanto maior os bits de cada chave maior o tempo que leva para fazer a tentativas de todos os elementos possiveis.
Aumentando seu carater exponencialmente em tempo conforme crescer o problema computacional. Dependendo o tamanho da chave ainda é rapido para computadores ou super maquinas.

SHA-256

O SHA-256 (Secure Hash Algorithm 256-bit) é um algoritmo de hash criptográfico projetado pela NSA que gera uma assinatura digital única e fixa de 256 bits (64 caracteres hexadecimais) para qualquer dado de entrada. Ele é irreversível, altamente seguro e amplamente utilizado para garantir a integridade de arquivos, assinaturas digitais e em blockchains.




```python
mensagem = "DADOS"
chave = 42

print("Criptografando...")
cifrado = []
for letra in mensagem:
    # ord() converte a letra para seu valor numérico na tabela ASCII
    # ^ é o operador lógico XOR em Python
    valor_cifrado = ord(letra) ^ chave
    cifrado.append(valor_cifrado)

print(f"Resultado cifrado (em números): {cifrado}")
```

```python
dados = [110, 107, 110, 101, 121]
chave = 42
print("desCriptografando...")
descifrado = []
for numero in dados:
  valor_descifrado = chr(numero ^ chave)
  descifrado.append(valor_descifrado)
print(f"Resultado descifrado (em letras): {descifrado}")
##Documente no seu relatório: Qual é a palavra original descoberta?
##Saida: desCriptografando...
##Resultado descifrado (em letras): ['D', 'A', 'D', 'O', 'S']

```

```python
import hashlib

texto1 = "Auditoria de Sistema"
texto2 = "Auditoria de Sistemas" # Uma letra 's' removida

hash1 = hashlib.sha256(texto1.encode()).hexdigest()
hash2 = hashlib.sha256(texto2.encode()).hexdigest()

hash1obj = hashlib.sha256(texto1.encode())
hash2obj = hashlib.sha256(texto2.encode())

print(f"Hash 1: {hash1}")
print(f"Hash 2: {hash2}\n")

print(f"Hash 1 Sem o hex: {hash1obj}")
print(f"Hash 2 sem o hex: {hash2obj}")


```
